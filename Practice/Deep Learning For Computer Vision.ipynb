```python
import numpy as np
from numpy import genfromtxt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import confusion_matrix, classification_report
from keras.models import Sequential
from keras.layers import Dense

data = genfromtxt("../Datasets/Bank Notes/bank_note_data.txt", delimiter=",")
print(data)
```

```python
#The labels are the last column of the matrix.

labels = data[:, 4]
print("labels: ", labels)

#Features are all the columns excluding the last one
features = data[:,0:4]
print("\nfeatures:\n", features)

X = features
y = labels

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
print("\n", X_train, "\n", "\nlen(X_train): ", len(X_train))

print("\n", X_test, "\n", "\nlen(X_test): ", len(X_test))

print("\n", y_train, "\n", "\nlen(X_test): ", len(y_train))

print("\n", y_test, "\n", "\nlen(y_test): ", len(y_test))
```

```python
scaler_object = MinMaxScaler()

scaler_object.fit(X_train)

scaled_X_train = scaler_object.transform(X_train)

scaled_X_test = scaler_object.transform(X_test)

print(scaled_X_train.max())

print(scaled_X_train.min())

model = Sequential()

model.add(Dense(4, input_dim=4, activation='relu'))

model.add(Dense(8, activation='relu'))

model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(scaled_X_train, y_train, epochs=50, verbose=2)
```

```python
print(model.metrics_names)

predictions = model.predict_classes(scaled_X_test)

confusion_matrix(y_test, predictions)

print(classification_report(y_test, predictions))
```

```python
import numpy as np
from numpy import genfromtxt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense
from sklearn.metrics import confusion_matrix, classification_report

# Load data from the file
data = genfromtxt("../Datasets/Bank Notes/bank_note_data.txt", delimiter=",")
print(data)

# The labels are the last column of the matrix.
labels = data[:, 4]
print("labels: ", labels)

# Features are all the columns excluding the last one
features = data[:, 0:4]
print("\nfeatures:\n", features)

X = features
y = labels

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
print("\n", X_train, "\n", "\nlen(X_train): ", len(X_train))

print("\n", X_test, "\n", "\nlen(X_test): ", len(X_test))

print("\n", y_train, "\n", "\nlen(y_train): ", len(y_train))

print("\n", y_test, "\n", "\nlen(y_test): ", len(y_test))

scaler_object = MinMaxScaler()

scaler_object.fit(X_train)

scaled_X_train = scaler_object.transform(X_train)

scaled_X_test = scaler_object.transform(X_test)

print(scaled_X_train.max())

print(scaled_X_train.min())

model = Sequential()

model.add(Dense(4, input_dim=4, activation='relu'))

model.add(Dense(8, activation='relu'))

model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(scaled_X_train, y_train, epochs=50, verbose=2)

print(model.metrics_names)

predictions = model.predict(scaled_X_test)
predictions_classes = (predictions > 0.5).astype(int)

conf_matrix = confusion_matrix(y_test, predictions_classes)
print(conf_matrix)

class_report = classification_report(y_test, predictions_classes)
print(class_report)
```
