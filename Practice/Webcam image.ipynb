```python
import cv2
import matplotlib.pyplot as plt

def main():
    cap = cv2.VideoCapture(0)
    
    ret, frame = cap.read()
    
    if cap.isOpened():
        ret, frame = cap.read()
        print(ret)
        
    img = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    
    
    plt.imshow(img)
    plt.show()
    cap.release()
    
if __name__ == "__main__":
    main()
```
