## Classification Vs Regression
<br>

<img src = "../Images/classification_regression.png"><br>

Both Classification and Regression models use various algorithms, such as logistic regression, decision trees, random forests, support vector machines (SVM), and neural networks.

However,

- In Classifcation, the model's output is a class label or probability distribution over the classes.
- In Regression, the model's output is a numerical value or a function that maps the input features to a continuous output.

<h3>1.) Classifcation</h3><br>
    
Classification is the task of predicting a categorical or discrete outcome variable based on input features. In classification, the target variable is divided into distinct classes or categories. The goal is to learn a model that can assign new, unseen data points to one of these predefined classes.

Examples of classifcation problems include:

- Predicting whether an email is spam or not.
- Identifying whether a transaction is fraudulent or legitimate.
- Classifying images of handwritten digits into their respective numbers (0-9).
- Determining the sentiment (positive, negative, neutral) of a text review.

<h3>2.) Regression</h3><br>

Regression is the task of predicting a continuous or numerical outcome variable based on input features. In regression, the target variable is a real-valued quantity that can take any numeric value within a given range. The goal is to learn a model that can estimate or approximate the relationship between the input features and the continuous target variable.

Examples of regression problems include:

- Predicting the price of a house based on its features (e.g., area, number of rooms).
- Forecasting the stock market index based on historical data and other factors.
- Estimating the age of a person based on demographic information.
- Predicting the temperature based on weather conditions.

## Overfitting and Underfitting
<br>

Overfitting and underfitting are common problems in machine learning models, and they can occur regardless of whether the data is split into training and testing sets. However, the use of training and testing sets can help in identifying and mitigating these problems.

<img src="../Images/overfitting_underfitting.png">

<h3>Overfitting</h3><br>

- Overfitting occurs when a model learns the training data too well, to the extent that it starts capturing noise or random variations in the data. The model becomes too complex and specific to the training data, resulting in poor generalization to unseen data. In other words, the model memorizes the training examples instead of learning the underlying patterns. Overfitting often leads to high training accuracy but low testing accuracy. It is a common problem when the model is too complex or when there is insufficient training data.

- Underfitting occurs when a model is too simple and fails to capture the underlying patterns and relationships in the data. The model is unable to learn the complexities present in the data, resulting in poor performance on both the training and testing data. Underfitting often leads to low training accuracy and low testing accuracy. It is a common problem when the model is too simple or when there is a lack of appropriate features or complexity to represent the data adequately.

```python
x = data.drop(['Species'], axis=1)
y = data['Species']
print(x.head())
print(x.shape)
print(y.head())
print(y.shape)
```

## Split the dataset into a training and a testing set
<br>

## Advantages 

<h3>1.) Training and Testing Separation</h3><br>

- The dataset can be splitted into two seperate sets: training and a testing set. <b> We can use the training set to build and train our model and then evaluate its performance on the testing set. </b>
- This ensures that we won't have the same observations in both sets.<b> Moreover, it also ensure that the model is tested on unseen data, providing a more realistic evaluation of its performance.</b>
- More flexible and faster than creating a model using all of the dataset for training.

<h3>2.) Avoiding Overfitting</h3><br>

- <b>Overfitting occurs when a model becomes too complex and starts to memorize the training data instead of learning general patterns.</b> By evaluating the model on unseen data, we can detect if the model is overfitting and making adjustments accordingly.

<h3>3.) Flexibility and Efficiency</h3><br>

- We can iterate and modify the model based on the performance on the testing set without the need to retrain the model on the entire dataset. This approach saves computational time and resources.

## Disadvantages
<br>

<h3>1.) Variance in Testing Set</h3><br>

- The accuracy scores obtained from the testing set may vary depending on the specific observations included in the set. If the testing set contains observations that are significantly different from the training set, the model's performance on the testign set may not accurately represent its generalization ability.

<h3>Additional notes:</h3><br>

- The accuracy of the model depends on the composition of the testing set, which is determined by the random seed used for splitting the data. Different splits can result in different accuracy scores.
- The complexity of the model plays a role in its performance. A too complex or too simple model may lead to lower testing accuracy.



```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.4)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

## Random Forest
<br>

<img src="../Images/random_forest.svg">

Random Forest is an ensemble learning method in machine learning that combines multiple decision trees to make more accurate predictions. It is a versatile and powerful algorithm that can be used for both classification and regression tasks.


<h2><b>Key concepts behind Random Forest:<br><br><br>1.) Ensemble Learning</b></h2><br>

Ensemble learning refers to the technique of combining multiple individual models to create a more robust and accurate model. Random Forest is an example of ensemble learning, where the individual models are decision trees.

<h2><b>2.) Decision Trees</b></h2><br>

A decision tree is a flowchart-like structure where each internal node represents a feature or attribute, each branch represents a decision rule, and each leaf node represents the outcome or prediction. Decision trees are popular models for making predictions because they are easy to understand and interpret.

<img src="../Images/decision_tree.png">

<h2><b>3.) Randomness</h2><br>
    
Random Forest introduces randomness in two ways.
    
- <b>Random Sampling:</b> Random Forest randomly selects a subset of the training data (with replacement) to build each decision tree. This process is known as "bootstrap aggregating" or "bragging". It helps to introduce diversity among the individual tree, making them less correlated.
    
- <b>Random Feature Selection:</b> At each node of the decison tree, Random Forest randomly selects a subset of features to consider whether making a split. This random feature selection further increases the diversity among the trees.
    
<h2><b>4.) Voting or Averaging</h2><br> Random Forest combines the predictions from multiple decision trees by using either voting (for classification tasks) or averaging (for regression tasks). Each decision tree in the forest independently predicts the outcome, and the final prediction is determined by majority voting (for classification) or averaging (for regression).

<h2><b>Advantages of Random Forest</h2><br>
    
- <b>High Accuracy:</b> Random Forest generally produces highly accurate predictions due to the combination of multiple decision trees and their independent predictions.
- <b>Robustness to Outliers and Noise:</b> Random Forest is robust to outliers and noisy data because it uses multiple decision trees, which are less likely to be affected by individual noisy or outlier data points.
- <b>Feature Importance:</b> Random Forest provides a measure of feature importance, indicating which features have the most significant impact on the prediction.
- <b>Versatlity:</b> Random Forest can handle a wide range of data types and can be applied to both classification and regression tasks.
- <b>Parallelization:</b> Training the decision trees in Random Forest can be easily parallelized, making it suitable for large datasets.

```python
from sklearn.ensemble import RandomForestClassifier

# Random Forest Classifier

"""
Number of decision trees = 100, the number of parallel jobs to run for both fitting and 
predicting is 1.
 
""" 
rf_model = RandomForestClassifier(n_estimators=100, n_jobs=1)
rf_model.fit(X_train, y_train)  # Fit the model with training data

# Predict
y_pred = rf_model.predict(X_test)
```
