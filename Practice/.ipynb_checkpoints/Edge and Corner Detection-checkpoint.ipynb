<h2><b>Edge and Corner Detection</b></h2><br>



```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

<h2><b>Image Gradient</b></h2><br>

```python
warehouse_tools = cv2.imread("../Images/warehouse_tools.jpg", 0)

plt.imshow(warehouse_tools, cmap = "gray")
plt.show()
```

```python
#Smooth the image

warehouse_tools_smoothen = cv2.GaussianBlur(src=warehouse_tools, ksize=(5,5), sigmaX=100, sigmaY=100)
plt.imshow(warehouse_tools_smoothen, cmap='gray')
plt.show()

warehouse_tools_smoothen2 = cv2.GaussianBlur(src=warehouse_tools_smoothen, ksize=(5,5), sigmaX=0, sigmaY=0)
plt.imshow(warehouse_tools_smoothen2, cmap='gray')
plt.show()
```

<h2><b>Sobel</b></h2><br>

```python
sobelx = cv2.Sobel(warehouse_tools,cv2.CV_8U,1,0,ksize=5)
sobely = cv2.Sobel(warehouse_tools,cv2.CV_8U,0,1,ksize=5)

# Display image
plt.imshow(sobelx,cmap='gray')
plt.show()

plt.imshow(sobely,cmap='gray')
plt.show()
```

```python
sobel_edge = cv2.bitwise_or(sobelx, sobely)

# Display image
plt.imshow(sobel_edge,cmap='gray')
plt.show()
```

```python
laplacian = cv2.Laplacian(warehouse_tools, cv2.CV_16S)  # Use CV_16S as the output data type

# Convert the output to a suitable range for display
laplacian = cv2.convertScaleAbs(laplacian)

# Display the image
plt.imshow(laplacian, cmap='gray')
plt.show()
```

<h2><b>Canny Edge Detector</b></h2><br>

```python
warehouse_tools_blurred = cv2.GaussianBlur(src=warehouse_tools, ksize=(5,5), sigmaX=0, sigmaY=0)
edges = cv2.Canny(warehouse_tools_blurred,100,150)
plt.imshow(edges,cmap='gray')
plt.show()

edges = cv2.Canny(warehouse_tools_blurred,100,150)
plt.imshow(edges,cmap='gray')
plt.show()
```

<h2><b>Harris Corner Detection</b></h2><br>

```python
img = cv2.imread('../Images/three_d.webp',0)

# Find the corners using the Harris Corner Detector
dst = cv2.cornerHarris(img,2,3,0.04)

# Display image
plt.imshow(dst,cmap ='gray')
plt.show()
```

```python
#result is dilated for marking the corners, not important
dst = cv2.dilate(dst,None)

# Display image
plt.imshow(dst,cmap ='gray')
plt.show()
```
