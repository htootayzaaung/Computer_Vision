<b>Read, Show and Write and Image</b>

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#Read Image as a grayscale

img1 = cv2.imread("../Images/anonymous_hacker.jpg", cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread("../Images/rainbow_six.jpg", cv2.IMREAD_GRAYSCALE)
print("img1:\n", img1)
print("Size: ", img1.shape)
print("Datatype: ", img1.dtype, "\n")

print("img2:\n", img2)
print("Size: ", img1.shape)
print("Datatype: ", img1.dtype)
```

```python
plt.imshow(img1)
```
