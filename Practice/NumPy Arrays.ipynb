```python
import numpy as np

mylist = [1, 2, 4]

print(type(mylist), "\n")

nparray = np.array(mylist)   #type-casting a list into numpy array

print(type(nparray), "\n")

nparray1 = np.arange(0, 10)

print(nparray1, "\n")

five_five_matrix = np.zeros(shape = (5, 5))

print(five_five_matrix, "\n")

ten_five_matrix = np.zeros(shape = (10, 5))

print(ten_five_matrix, "\n")

print(type(0), "\n")

print(type(0.), "\n")

ones_2x4 = np.ones((2, 4))

print(ones_2x4, "\n")

ones_2x4 = np.ones(shape = (2, 4))

print(ones_2x4, "\n")
```

```python
np.random.seed(101)
random_arr = np.random.randint(0, 100, 10)   #0 to 100 with 10 elements
print("Array: ", random_arr, "\n")
print("Largest number in the array: ", random_arr.max(), "\n")
print("Smallest number in the array: ", random_arr.min(), "\n")
print("Index of the largest number: ", random_arr.argmax(), "\n")
print("Index of the smallest number: ", random_arr.argmin(), "\n")
print("Mean of the numbers in the array: ", random_arr.mean(), "\n")
print(random_arr.reshape(2, 5), "\n")

mat = np.arange(0, 100).reshape(10, 10)
print(mat, "\n")

for i in range(10):
    for j in range(10):
        print(mat[i][j], end=" ")
    print("\n")

print("\n")

for x in range (10):
    for y in range(10):
        print(mat[x, y], end=" ")
    print("\n")

print("\n")
```

```python
import matplotlib.pyplot as plt
from PIL import Image
%matplotlib inline

pic = Image.open("../Images/eiffel2.jpg")
print(type(pic))
plt.imshow(pic)
plt.axis("off")
plt.show()

pic_nparray = np.asarray(pic)
print(type(pic_nparray), "\n", pic_nparray.shape)
plt.imshow(pic_nparray)
plt.axis("off")
plt.show()

pic_red = pic.copy()
plt.imshow(pic_red)
plt.axis("off")
plt.show()


pic_red_nparray = np.asarray(pic_nparray[:, :, 0])
plt.imshow(pic_red_nparray)
plt.axis("off")
plt.show()

plt.imshow(pic_red_nparray, cmap='gray')
plt.axis("off")
plt.show()

```
