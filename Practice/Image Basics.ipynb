```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import cv2
```

```python
img = cv2.imread("../Images/eiffel2.jpg")

print(type(img))
plt.imshow(img)
plt.show()
```

```python
correct_color_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(correct_color_img)
plt.show()
```

```python
grayscale_img =cv2.cvtColor(img, cv2.IMREAD_GRAYSCALE)
plt.imshow(grayscale_img, cmap='gray')
plt.show()
```

```python
print(correct_color_img.shape)
resized_color_img = cv2.resize(correct_color_img, (350, 600))
plt.imshow(resized_color_img)
```

## Reducing a size of an image by a scale-factor
<br>

```python
w_ratio = 0.5
h_ratio = 0.5

new_img = cv2.resize(correct_color_img, (0, 0), correct_color_img, w_ratio, h_ratio)
plt.imshow(new_img)
plt.show()
```

## Flipping images
<br>

```python
flip1 = cv2.flip(new_img, -1)
plt.imshow(flip1)
plt.show()

flip2 = cv2.flip(new_img, 1)
plt.imshow(flip2)
plt.show()

flip3 = cv2.flip(new_img, 0)
plt.imshow(flip3)
plt.show()

```

```python
def draw_circle(event, x, y, flags, param):
    
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(img, (x, y), 100, (0, 255, 0), -1)
        
    elif event == cv2.EVENT_RBUTTONDOWN:
        cv2.circle(img, (x, y), 100, (255, 0, 0), -1)
        
cv2.namedWindow(winname='my_drawing')

cv2.setMouseCallback('my_drawing', draw_circle)

img = np.zeros((512, 512, 3), np.int8)

while True:
    
    cv2.imshow('my_drawing', img)
    
    if cv2.waitKey(20) & 0xFF == 27:
        break
        
cv2.destroyAllWindows()
```

```python
drawing = False   #True while mouse button down, False while mouse button UP
ix, iy = -1, -1

def draw_rectangle(event, x, y, flags, params):
    global ix, iy, drawing
    
    if event == cv2.EVENT_LBUTTONDOWN:
        
        drawing = True
        ix, iy = x, y
        
    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing == True:
            cv2.rectangle(img, (ix, iy), (x, y), (0, 255, 0), -1)
            
    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
        cv2.rectangle(img, (ix, iy), (x, y), (0, 255, 0), -1)
        
img = np.zeros((512, 512, 3))

cv2.namedWindow(winname='my_drawing')

cv2.setMouseCallback('my_drawing', draw_rectangle)


while True:
    
    cv2.imshow('my_drawing', img)
    
    if cv2.waitKey(1) & 0xFF == 27:
        break

cv2.destroyAllWindows()
```
